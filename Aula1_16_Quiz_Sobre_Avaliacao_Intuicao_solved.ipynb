```python
# Importe o pandas
import pandas as pd

# Carregue os dados de renda do censo
df_census = pd.read_csv("census_income_data.csv")

# Imprimindo os iniciais para conferir
display(df_census.head())
```

```python
# Para respondermos algumas perguntas, o método describe irá ajudar
#Uso display ao invés de print pois o formato fica melhor no notebook
display (df_census.describe())
```

```python
# Vamos armazenar o resultado da describe para não chamá-la a todo momento
describe = df_census.describe()

# N de linhas:
print("Número de linhas usando describe: {0}".format(describe['age']['count']))
print("Número de linhas usando shape: {0} \n".format(df_census.shape[0]))

# N de colunas:
print("Número de colunas: {0} \n".format(df_census.shape[1]))

# Tipos de dados de várias colunas (\n pula linha)
print("Tipo de age: {0}".format(df_census.age.dtype))
print("Tipo de workclass: {0}".format(type(df_census.workclass[0])))
print("Tipo de education: {0}".format(type(df_census.education[0])))
print("Tipo de relationship: {0}".format(type(df_census.relationship[0])))
print("Tipo de hours-per-week: {0}".format(type(df_census["hours-per-week"][0])))
print("Tipo de income: {0} \n".format(type(df_census.income[0])))
# Valores Faltantes (poderíamos verificar uma coluna por vez mas 
#  essa forma é elegante)
print("Colunas com valores Faltantes: \n {0} \n".format(
    df_census.columns[df_census.isnull().any()].values))

# N de valores únicos de educação
print("Valores Únicos de educação: {0}".format(df_census["education"].nunique()))

# Idade Média
print("Idade média: {0} \n".format(round(describe["age"]["mean"], 0)))

# 75% horas por semana
print("75% horas por semana: {0}".format(describe["hours-per-week"]["75%"]))
```
